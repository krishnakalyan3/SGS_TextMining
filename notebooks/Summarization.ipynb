```python
from gensim.summarization import summarize
```

```python
import os
#CORPUS_PATH = os.path.join('data/sample/002/')
CORPUS_PATH = os.path.join('data/001')
filenames1 = sorted([os.path.join(CORPUS_PATH, fn) for fn in os.listdir(CORPUS_PATH)])
CORPUS_PATH2 = os.path.join('data/002')
filenames2 = sorted([os.path.join(CORPUS_PATH2, fn) for fn in os.listdir(CORPUS_PATH2)])
```

```python
def doc_read(PATH):
    with open(PATH, "r", encoding='UTF-8') as f:
        i = f.read()
    return i
```

```python
read_data1 = []
for i in filenames1:
    k = doc_read(i)
    read_data1.append(k)
```

```python
read_data2 = []
for i in filenames2:
    k = doc_read(i)
    read_data2.append(k)
```

```python
def unique_word_count(data):
    wc_dict = {}
    data_break = data.split(" ")
    for i in data_break:
        if i not in wc_dict:
            wc_dict[i] = 1
        else:
            wc_dict[i] += 1
    return wc_dict
```

```python
read_total = read_data1 + read_data2
```

```python
ARTICLE_INDEX = 1003
print(filenames1[ARTICLE_INDEX])
original_doc = doc_read(filenames1[ARTICLE_INDEX])
```

```python
original_doc
```

```python
word_count = min(20+0.1*len(unique_word_count(original_doc)), 80)
print(word_count)
summary = summarize(read_total[ARTICLE_INDEX], word_count=word_count)
summary
```
